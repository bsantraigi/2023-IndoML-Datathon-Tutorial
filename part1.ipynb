## Visualization

```python
# Let's construct a dataframe from the dataset for easier visualization
import pandas as pd

df = pd.DataFrame(dataset['train'])
```

```python
df
```

```python
# Lets do some VISUALIZATION!

# You can also use matplotlib for this, but seaborn has better defaults and it uses matplotlib under the hood.

# For more examples of seaborn/matplotlib plots, see 
# 1. https://seaborn.pydata.org/examples/index.html
# 2. https://python-graph-gallery.com/
import seaborn as sns
```

```python
# Let's plot the distribution of the number of locales in the dataset.
## Rotate x axis labels to make them readable and increase figure size

sns.set(rc={'figure.figsize':(14, 9)}) # Setting some matplotlib configs

_ = sns.countplot(x='locale', data=df)\
    .set_xticklabels(
        sns.countplot(x='locale', data=df).\
            get_xticklabels(), 
            rotation=90
            )
```

```python
# That is quite well balanced!
# Tips: For the purpose of this competition, you can simulate 
# few-shot learning by using only a subset of the data. 
# You can create new train/validation/test splits by based on the domains/locales.

# Next, let's plot the distribution of the number of intents in the dataset.
_ = sns.countplot(x='intent', data=df)\
    .set_xticklabels(
        sns.countplot(x='intent', data=df).\
            get_xticklabels(),
            rotation=90
            )
```

```python
# Ok, now that's an interesting distribution.
# Let's take a look at that in sorted order, 
# and also convert the y-axis to percentage and map the ids to the actual intent names.
import numpy as np

# A detail about huggingface datasets API: 
# ClassLabel: Used for storing categorical features or labels.
# e.g. dataset['train'].features['intent'] is a ClassLabel 
# and we can use .int2str(...) method to map the ids to the actual intent names.

intent_counts = df['intent'].value_counts()
intent_counts = intent_counts.reset_index()
intent_counts.columns = ['intent', 'count']
intent_counts['percent'] = intent_counts['count'] / intent_counts['count'].sum()
intent_counts['intent'] = intent_counts['intent'].map(dataset['train'].features['intent'].int2str)

_ = sns.barplot(x='intent', y='percent', data=intent_counts.sort_values(by='percent', ascending=False))\
    .set_xticklabels(
        sns.barplot(x='intent', y='percent', data=intent_counts.sort_values(by='percent', ascending=False)).\
            get_xticklabels(),
            rotation=90
            )
```

```python
# TIP: So, as you can guess, we will have to take notice of the 
# class imbalance while training our models. Also, during 
# evaluation, we should use macro-averaged F1 score to ensure
# that we are not overfitting towards the majority classes.
```
